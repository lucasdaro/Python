```python
import pandas as pd
dados = '/content/Media_precos.csv'
```

```python
df = pd.read_csv(dados, sep=';', encoding='utf-8')
```

```python
gasolina = df['Produto'] == 'GASOLINA'
```

```python
df_gasolina = df[gasolina]
```

```python
import matplotlib.pyplot as plt
import numpy as np
df_gasolina
```

```python
anos = df_gasolina['Ano']
precos = df_gasolina['Valor de Venda']
```

```python
fig, ax = plt.subplots()
```

```python
fig, ax = plt.subplots()
ax.plot(anos,precos, linewidth=2.0)
```

```python
#
dados_gasolina = [[df_gasolina['Ano']],[df_gasolina['Valor de Venda']]]
dados_gasolina
```

```python
fig, ax = plt.subplots()
ax.plot(anos,precos, linewidth=2.0)
```

```python
ano_min = df_gasolina['Ano'].min()
ano_max = df_gasolina['Ano'].max()
```

```python
anos_do_grafico = np.arange(ano_min,ano_max+0.5, 0.5)
anos_do_grafico
```

```python
preco_gasolina = df_gasolina['Valor de Venda']
preco_gasolina
```

```python
fig, ax = plt.subplots(figsize=(14,10))
ax.set_xticks(anos_do_grafico)
ax.tick_params(axis='x', rotation=45)

ax.set_title('Preo Combustivel')
ax.set_xlabel('Ano')
ax.set_ylabel('Preos')
ax.legend()

ax.plot(anos_do_grafico, preco_gasolina, label='Gasolina', color='yellow')
```

```python
etanol = df['Produto'] == 'ETANOL'
tabela_etanol = df[etanol]
preco_etanol = tabela_etanol['Valor de Venda']
preco_etanol
```

```python
fig, ax = plt.subplots(figsize=(14,10))
ax.set_xticks(anos_do_grafico)
ax.tick_params(axis='x', rotation=45)

ax.set_title('Preo Combustivel')
ax.set_xlabel('Ano')
ax.set_ylabel('Preos')

ax.plot(anos_do_grafico, preco_etanol, label='Etanol', color='green')
ax.plot(anos_do_grafico, preco_gasolina, label='Gasolina', color='yellow')

ax.legend()
```

```python
diesel = df['Produto'] == 'DIESEL'
tabela_diesel = df[diesel]
preco_diesel = tabela_diesel['Valor de Venda']
preco_diesel
```

```python
fig, ax = plt.subplots(figsize=(14,10))
ax.set_xticks(anos_do_grafico)
ax.tick_params(axis='x', rotation=45)

ax.set_title('Preo Combustivel')
ax.set_xlabel('Ano')
ax.set_ylabel('Preos')

ax.plot(anos_do_grafico, preco_etanol, label='Etanol', color='green')
ax.plot(anos_do_grafico, preco_diesel, label='Diesel', color='red')
ax.plot(anos_do_grafico, preco_gasolina, label='Gasolina', color='yellow')

ax.legend()
```

```python
gas = df['Produto'] == 'GNV'
tabela_gas = df[gas]
preco_gas = tabela_gas['Valor de Venda']
preco_gas
```

```python
fig, ax = plt.subplots(figsize=(14,10))
ax.set_xticks(anos_do_grafico)
ax.tick_params(axis='x', rotation=45)

ax.set_title('Preo Combustivel')
ax.set_xlabel('Ano')
ax.set_ylabel('Preos')

ax.plot(anos_do_grafico, preco_etanol, label='Etanol', color='green')
ax.plot(anos_do_grafico, preco_gas, label='Gas', color='blue')
ax.plot(anos_do_grafico, preco_diesel, label='Diesel', color='red')
ax.plot(anos_do_grafico, preco_gasolina, label='Gasolina', color='yellow')

ax.legend()
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(14,10))

ax.set_xticks(anos_do_grafico)
ax.tick_params(axis='x', labelsize=12, rotation=45)
ax.tick_params(axis='y', labelsize=12)

ax.set_title('Preo Combustvel', fontsize=16)
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Preos', fontsize=14)

ax.plot(anos_do_grafico, preco_etanol, label='Etanol', color='green', linewidth=2, marker='o')
ax.plot(anos_do_grafico, preco_gas, label='Gas', color='blue', linewidth=2, marker='o')
ax.plot(anos_do_grafico, preco_diesel, label='Diesel', color='red', linewidth=2, marker='o')
ax.plot(anos_do_grafico, preco_gasolina, label='Gasolina', color='yellow', linewidth=2, marker='o')

ax.grid(True)
ax.legend(loc='best', fontsize=12)

plt.show()
```
